```python
import folium
from hwnetwork import highways
```

```python
hw = highways()
point = [43.5, -112.5]
```

```python
def mapClosestWay(pt=point, dir='start'):
    pts = [pt]
    way = hw.findClosestWay(point[0], point[1], dir)
    pts.append([way['startNode']['lat'], way['startNode']['long']])
    pts.append([way['endNode']['lat'], way['endNode']['long']])

    m = folium.Map()
    points = pts[1:]
    name = way['tags']['ref'] if 'ref' in way['tags'] else ''
    folium.PolyLine(points, weight=5, opacity=1, popup=name).add_to(m)
    for i in [way['startNode'], way['endNode']]:
        folium.Marker(location = [i['lat'], i['long']]).add_to(m)
    folium.Marker(location = pt, popup = f'lat: {pt[0]}\nlon: {pt[1]}').add_to(m)
    lats = [i[0] for i in pts]
    lons = [i[1] for i in pts]
    m.fit_bounds([[min(lats), min(lons)], [max(lats), max(lons)]])
    return m
```

```python
m = mapClosestWay(point, 'start')
m
```
